```python

import numpy as np

# Training data
X = np.array([[0,0],[0,1],[1,0],[1,1]])
T = np.array([0,0,0,1]) #AND Gatew
#For OR Gate set T = np.array([0,1,1,1])
#For XOR Gate set T = np.array([0,1,1,0])

w = np.zeros(2)
b = 0
lr = 1

def step(v):
    return 1 if v >= 0 else 0

# Training
for epoch in range(10):
    for i in range(len(X)):
        v = np.dot(X[i], w) + b
        y = step(v)
        e = T[i] - y
        
        w = w + lr * e * X[i]
        b = b + lr * e

print("Weights:", w)
print("Bias:", b)
```
