```python
import numpy as np

# inputs
x = np.array([1, 0, 1])
w = np.array([0.6, 0.4, 0.8])
b = -0.5

# weighted sum
z = np.dot(x, w) + b

# activation (step function)
output = 1 if z >= 0 else 0

print("Weighted sum:", z)
print("Neuron output:", output)
```

```python
def mc_pitts_and(x1, x2):
    w1, w2 = 1, 1
    threshold = 2
    z = x1*w1 + x2*w2
    return 1 if z >= threshold else 0

for x1 in [0,1]:
    for x2 in [0,1]:
        print(x1, x2, "→", mc_pitts_and(x1, x2))
```

```python
import numpy as np

# training data (AND gate)
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,0,0,1])

w = np.zeros(2)
b = 0
lr = 0.1

for epoch in range(10):
    for i in range(len(X)):
        z = np.dot(X[i], w) + b
        y_pred = 1 if z >= 0 else 0
        error = y[i] - y_pred
        w += lr * error * X[i]
        b += lr * error

print("Weights:", w)
print("Bias:", b)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

sigmoid = 1 / (1 + np.exp(-x))
relu = np.maximum(0, x)
tanh = np.tanh(x)

plt.plot(x, sigmoid, label="Sigmoid")
plt.plot(x, relu, label="ReLU")
plt.plot(x, tanh, label="Tanh")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

# input features
x = np.array([1, 1, 1, 1])  # features detected

# second layer weights (digit '2')
w = np.array([0.8, 0.9, 0.7, 0.8])
b = -2.5

z = np.dot(x, w) + b
output = 1 / (1 + np.exp(-z))

print("Probability of digit 2:", output)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

data = load_iris()
X, y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X, y)

model = MLPClassifier(hidden_layer_sizes=(5,), max_iter=1000)
model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```

```python
from sklearn.datasets import load_digits
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

digits = load_digits()
X_train, X_test, y_train, y_test = train_test_split(
    digits.data, digits.target)

clf = MLPClassifier(hidden_layer_sizes=(64,), max_iter=1000)
clf.fit(X_train, y_train)

print("Digit recognition accuracy:", clf.score(X_test, y_test))

```
