```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
# read the input data
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/pu9kbeSaAtRZ7RxdJKX9_A/yellow-tripdata.csv'
raw_data = pd.read_csv(url)
raw_data

corr_values = raw_data.corr()['tip_amount'].drop('tip_amount')
loweffect = ['improvement_surcharge','store_and_fwd_flag','payment_type','VendorID']
corr_values = corr_values.drop(loweffect)
corr_values.plot(kind = 'barh',figsize=(10,6))
```

```python
y = raw_data['tip_amount'].astype('float32')
features = raw_data.drop('tip_amount',axis = 1)
X = features.values
X = normalize(X,axis = 1, norm = 'l1',copy = False)

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=69,test_size=0.2)
```

```python
model = DecisionTreeRegressor(max_depth = 5,random_state = 72)
model.fit(X_train,y_train)

y_predict = model.predict(X_test)
mse_score = mean_squared_error(y_test,y_predict)
mse_score
```
