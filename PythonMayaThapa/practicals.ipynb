```python

def power_set(nums):
    if not nums:
        return [[]]
    smaller = power_set(nums[1:])
    return smaller + [[nums[0]] + s for s in smaller]

l1= [1, 2, 3,4]

# Step 1: Generate all subsets
subsets = power_set(l1)

# Step 2: Keep only subsets with even sum
even_subsets = [s for s in subsets if sum(s) % 2 == 0]

# Step 3: Sort so smallest groups come last
even_subsets.sort(key=len, reverse=True)

# Print result
for group in even_subsets:
    print(group)
```
