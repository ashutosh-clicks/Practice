```python
#functions
from functools import reduce

a = [1,2,3,4,5,6]
for i,v in enumerate(a,10):
    print(f"{i} : {v}")

def gt5(num):
    if num>5:
        return num
gt5l = list(filter(gt5,a))

def sq(num):
    return num**2

asl = list(map(sq,a))
asl

def suma(num1,num2):
    return num1 +num2

asuma = reduce(suma,a)
asuma
```

```python
def factorial(n):
    return 1 if n <= 1 else n*factorial(n-1)

factorial(5)

```
